# 🔧 Instalações e Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

# 📥 Carregamento da Base de Dados

```python
df_base = pd.read_csv("base_unificada.csv")
df_base.head()
```

# 🔍 Diagnóstico Inicial dos Dados

Verificar o tamanho da base

```python
df_base.shape
```

Verificar quais Colunas Possuem 100% ou quase de Dados em Branco

```python
(df_base.isnull().sum() / len(df_base) * 100).sort_values(ascending=False)
```

# 🧹 Tratamento e Limpeza de Dados

Remover Colunas com 100% de Dados Nulos

```python
df_base = df_base.dropna(axis=1, how='all')
df_base.head()
```

Verificar quais Colunas possuem os Dados 100% identicos a outras

```python
# Verifica colunas duplicadas em conteúdo
duplicadas = []
for i in range(len(df_base.columns)):
    for j in range(i + 1, len(df_base.columns)):
        if df_base[df_base.columns[i]].equals(df_base[df_base.columns[j]]):
            duplicadas.append((df_base.columns[i], df_base.columns[j]))

# Exibe os pares de colunas duplicadas
print("Colunas com dados duplicados:")
for par in duplicadas:
    print(par)
```

Remover Colunas com Dados Duplicados

```python
colunas_para_remover = [
    'codigo_profissional',
    'telefone_x',
    'titulo_profissional',
    'email.1',
    'nome_y.1',
    'data_ultima_promocao'
]

df_base = df_base.drop(columns=colunas_para_remover)
df_base.head()
```

Verificar Nome das Colunas

```python
for col in df_base.columns:
    print(col)
```

Remover colunas que não serão necessárias para a nossa análise exploratória

```python
colunas_para_remover = [
    'prospects_id', 'applicant_id', 'codigo',
    'nome_x', 'nome_y', 'nome', 'cv_pt', ''
    'email', 'telefone_celular', 'telefone_y',
    'endereco', 'bairro', 'regiao',
    'data_aceite', 'data_criacao', 'data_atualizacao','fonte_indicacao',
    'download_cv', 'linkedin', 'id_ibrati',
    'vagas_id', 'vaga_sap', 'cliente_x', 'cliente_y',
    'analista_responsavel', 'empresa_divisao', 'requisitante',
    'solicitante_cliente', 'objetivo_vaga', 'prioridade_vaga',
    'origem_vaga', 'superior_imediato', 'nome_substituto',
    'equipamentos_necessarios', 'viagens_requeridas',
    'habilidades_comportamentais_necessarias',
    'demais_observacoes', 'local', 'unidade',
    'nivel_academico_y', 'nivel_ingles_y', 'nivel_espanhol_y',
    'outro_idioma_y', 'nivel profissional',
    'vagas_id', 'limite_esperado_para_contratacao',
    'data_final', 'data_inicial',
    'comentario', 'estado_civil',
    'conhecimentos_tecnicos', 'certificacoes', 'outras_certificacoes',
    'remuneracao', 'nivel_ingles_x', 'nivel_espanhol_x', 'outro_idioma_x',
    'instituicao_ensino_superior', 'cursos', 'ano_conclusao',
    'outro_curso', 'cargo_atual', 'data_admissao', 'data_requicisao',
    'titulo_vaga', 'tipo_contratacao', 'prazo_contratacao', 'pais',
    'local_trabalho', 'vaga_especifica_para_pcd', 'horario_trabalho', 'areas_atuacao',
    'principais_atividades', 'competencia_tecnicas_e_comportamentais',
    'valor_venda', 'valor_compra_1', 'valor_compra_2',
]

df_base = df_base.drop(columns=[col for col in colunas_para_remover if col in df_base.columns])
```

```python
valores_positivos = [
    'Contratado pela Decision',
    'Contratado como Hunting',
    'Aprovado',
    'Documentação CLT',
    'Documentação PJ'
]

df_base['situacao_candidado_'] = df_base['situacao_candidado'].apply(lambda x: 1 if x in valores_positivos else 0)

df_base.head()
```

```python
df_base.shape
```

Transformar os dados de acordo com o seu tipo

```python
#Transformação das Datas

df_base["data_candidatura"] = pd.to_datetime(df_base["data_candidatura"], errors="coerce", dayfirst=True)
df_base["ultima_atualizacao"] = pd.to_datetime(df_base["ultima_atualizacao"], errors="coerce", dayfirst=True)
df_base["data_nascimento"] = pd.to_datetime(df_base["data_nascimento"], errors="coerce", dayfirst=True)
```

```python
#Transformação em Categorias
colunas_categoricas = [
    "sexo", "estado_civil", "pcd", "modalidade", "nivel_profissional",
    "nivel_academico_x", "nivel_ingles_x", "nivel_espanhol_x", "outro_idioma_x",
    "area_atuacao", "recrutador", "situacao_candidado"
]

for col in colunas_categoricas:
    if col in df_base.columns:
        df_base[col] = df_base[col].astype("category")
```

```python
#Criação da variável binária foi_contratado
df_base["foi_contratado"] = (df_base["situacao_candidado"] == "Contratado pela Decision").astype(int)
```

```python
df_base.describe
```

```python
# Exportar o arquivo tratato para ser utilizado no streamlit

nome_arquivo_csv = 'base_tratada.csv'
df_base.to_csv(nome_arquivo_csv, index=False)

print(f"Base salva com sucesso como '{nome_arquivo_csv}' no diretório atual.")
```

# 📊 Análise Exploratória

```python
# Conta e ordena os status (maior para menor)
status_counts = df_base["situacao_candidado"].value_counts().sort_values(ascending=True)

# Plotagem
plt.figure(figsize=(14, 10))
bars = plt.barh(status_counts.index, status_counts.values, color='mediumpurple')

# Adiciona os valores formatados com ponto como separador de milhar
for bar, value in zip(bars, status_counts.values):
    valor_formatado = f"{value:,.0f}".replace(",", ".")
    plt.text(value + 100, bar.get_y() + bar.get_height() / 2,
             valor_formatado, va='center', ha='left', fontsize=9, fontweight='bold')

plt.xlabel("Quantidade")
plt.ylabel("Status")
plt.title("Distribuição dos Status dos Candidatos")
plt.tight_layout()
plt.show()
```

```python
# Filtrar apenas candidatos contratados
contratados = df_base[df_base["foi_contratado"] == True]

# Contar contratações por recrutador
contratacoes_recrutador = contratados["recrutador"].value_counts().sort_values(ascending=True)

# Plotar gráfico horizontal
plt.figure(figsize=(16, 18))
bars = plt.barh(contratacoes_recrutador.index, contratacoes_recrutador.values, color='mediumpurple')

# Adicionar os valores nas barras
for bar, value in zip(bars, contratacoes_recrutador.values):
    texto = f"{value:,}".replace(",", ".")
    plt.text(value + 1, bar.get_y() + bar.get_height() / 2,
             texto, va='center', ha='left', fontsize=9, fontweight='bold')

plt.title("Contratações por Recrutador", fontsize=14, fontweight="bold")
plt.xlabel("Quantidade de Candidatos Contratados")
plt.ylabel("Recrutador")
plt.tight_layout()
plt.show()
```

```python
# Calcula a taxa de conversão por recrutador
total_candidatos = df_base['recrutador'].value_counts()
contratados = df_base[df_base['situacao_candidado'] == 'Contratado pela Decision']['recrutador'].value_counts()

taxa_conversao = (contratados / total_candidatos * 100).dropna().sort_values(ascending=True)

# Define altura proporcional ao número de recrutadores
altura = len(taxa_conversao) * 0.3  # ajuste conforme necessidade

plt.figure(figsize=(16, altura))  # largura fixa, altura dinâmica

bars = plt.barh(taxa_conversao.index, taxa_conversao.values, color="mediumpurple")

# Adiciona os valores
for bar, value in zip(bars, taxa_conversao.values):
    plt.text(value + 1, bar.get_y() + bar.get_height() / 2,
             f"{value:.2f}%", va='center', ha='left', fontsize=7)

# Eixos e título
plt.xlabel("Taxa de Conversão (%)")
plt.ylabel("Recrutador")
plt.title("Taxa de Conversão por Recrutador", fontsize=14, fontweight="bold")

# Limites e layout
plt.xlim(0, 105)
plt.xticks(fontsize=9)
plt.yticks(fontsize=7)  # deixa mais compacto
plt.tight_layout(pad=2)
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.show()
```
